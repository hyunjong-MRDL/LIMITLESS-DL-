```python
import data
from utils import seed_everything, CFG

seed_everything(CFG["SEED"])

root = root = "D:/Datasets/SJS/"
processed_root = f"{root}Processed/"

total_SJS = data.split_by_patient(data.load_by_diagnosis(processed_root, "SJS"))
total_CTR = data.split_by_patient(data.load_by_diagnosis(processed_root, "NORMAL"))

train_SJS, test_SJS = data.train_test_split(total_SJS, 0.5)
train_CTR, test_CTR = data.train_test_split(total_CTR, 0.5)

train_X_pg, train_X_sg, train_y_pg, train_y_sg = data.gland_dataset(train_SJS, train_CTR)
test_X_pg, test_X_sg, test_y_pg, test_y_sg = data.gland_dataset(test_SJS, test_CTR)
```

```python
train_ID = list(train_SJS.keys())
test_ID = list(test_SJS.keys())

print(len(train_ID))
print(train_ID[:5])
print(len(test_ID))
print(test_ID[:5])
```

```python
import torch, torchvision, PIL, numpy

trans = torchvision.transforms.Compose([
    torchvision.transforms.Resize((256, 256)),
    torchvision.transforms.ToTensor()
])

pg_IDs = list(train_X_pg.keys())
sample_image = PIL.Image.open(train_X_pg[pg_IDs[0]][0])
print(sample_image.size)
sample = trans(sample_image)
print(sample.shape)
```

```python
import torch.nn as nn
pg_layer = nn.Sequential(
    nn.Conv2d(3, 32, 3, 1, 1),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(32, 64, 3, 1, 1),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(64, 128, 3, 1, 1),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(128, 256, 3, 1, 1),
    nn.ReLU(),
    nn.MaxPool2d(2)
)
sg_layer = nn.Sequential(
    nn.Conv2d(3, 32, 3, 1, 1),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(32, 64, 3, 1, 1),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(64, 128, 3, 1, 1),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(128, 256, 3, 1, 1),
    nn.ReLU(),
    nn.MaxPool2d(2),
)
fc_layer = nn.Linear(256 * 16 * 16, 1024)

fusion_layer = nn.Sequential(
    nn.Linear(1024, 256),
    nn.ReLU(),
    nn.Linear(256, 1),
    nn.Sigmoid()
)

pg_out = pg_layer(sample.unsqueeze(0))
sg_out = sg_layer(sample.unsqueeze(0))
print(pg_out.shape, sg_out.shape)
pg_out = fc_layer(pg_out.view(pg_out.size(0), -1))
sg_out = fc_layer(sg_out.view(sg_out.size(0), -1))
print(pg_out.shape, sg_out.shape)
fused = pg_out + sg_out
print(fused.shape)
out = fusion_layer(fused)
out[0]
```
